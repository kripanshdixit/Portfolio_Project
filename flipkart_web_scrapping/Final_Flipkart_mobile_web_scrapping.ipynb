# scrapping data from all the web_pages

```python
data_list=[]
def info_scrape(container_page):    
    for i in container_page:
        micro_list=[]
        try:
            micro_list.append(i.find('div' , class_ = '_4rR01T').text)
            micro_list.append(i.find('div' , class_ = '_3LWZlK').text)
            micro_list.append(i.find('span' , class_ = '_2_R_DZ').text.strip())
            micro_list.append(i.find('div' , class_ = '_30jeq3 _1_WHN1').text.strip()) # offer_price
            # market_price
            micro_list.append(i.find('div' , class_ = '_3I9_wc _27UcVY').text.strip())
            #discount
            micro_list.append(i.find_all('div' , class_ = '_3Ay6Sb')[0].text.strip())
        except:
            micro_list.append(None)

        data_list.append(micro_list)
```

```python
fi='https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1'
page_list.insert(0,fi)
```

```python
len(page_list)
```

```python
data_list=[]
for i in page_list:
    res=requests.get(i)
    data=BeautifulSoup(res.content,'html.parser')
    container_page=data.find_all('div' , class_ = '_1YokD2 _2GoDe3')[0].find_all('div' , class_ = '_1AtVbE col-12-12')
    info_scrape(container_page)
    
```
